```python
#this notebook instructs users on how to open and explore the large hdf5 files 

import h5py

#  STEP 1: Set your file path 
#change region name to be void, normal, rarepeak
file_path = "practice_halo_catalog.hdf5"

#  STEP 2: Open the file 
with h5py.File(file_path, "r") as f:

    print("\nTop-level keys in file:")
    print(list(f.keys()))

    #  STEP 3: Look at one halo -> halo names are top level keys 
    halo = f['halo_0']

    print("\nDatasets inside halo_0:")
    print(list(halo.keys()))

    print("\nAttributes in halo_0:")
    print(list(halo.attrs.keys()))

    #  STEP 4: Access a specific attribute 
    redshift = halo.attrs['current_redshift']
    print("\nRedshift of halo_0:", redshift)
```
